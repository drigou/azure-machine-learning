# Create a compute instance

Reference documentation for creating a compute instance through SDK: https://learn.microsoft.com/en-us/python/api/azure-ai-ml/azure.ai.ml.entities.computeinstance?view=azure-python

````python
from azure.ai.ml.entities import ComputeInstance
from azure.ai.ml import MLClient
from azure.identity import DefaultAzureCredential
import dotenv
import os

dotenv.load_dotenv(override=True)

ml_client = ml_client = MLClient(
     credential         = DefaultAzureCredential()
    ,subscription_id    = os.environ["AZURE_SUBSCRIPTION_ID"]
    ,resource_group_name= os.environ["AZURE_RESOURCE_GROUP"]
    ,workspace_name     = os.environ["AZURE_ML"]
)

ci_name = "basic-ci-1234a"
ci = ComputeInstance(
     name           = ci_name
    ,size           = "STANDARD_DS3_v2"
    ,description    = "Create test VM with SDK"
    ,idle_time_before_shutdown_minutes = 20
    # ,setup_scripts = ""
    )

ml_client.begin_create_or_update(ci).result()
````

````python
try:
    poller = ml_client.compute.begin_delete(ci_name)
    while not poller.done:
        print(f"Deleting compute: {ci_name}.", end="\r")
    print({f"Compute {ci_name} deleted"})
except Exception as e:
    raise e
````

## Create and use a compute cluster

- ```tier```: Specifies whether your virtual machines are low priority or dedicated. Setting to low priority can lower costs as you're not guaranteed availability.
- ```size```: Specifies the virtual machine type of each node within the compute cluster. Next to size, you can also specify whether you want to use CPUs or GPUs. You need to select a VM family that contains NVIDIA GPUs (typically the NC, ND, or NV series).

````python
from azure.ai.ml.entities import AmlCompute

cc_name = "cpu-cluster"

cc = AmlCompute(
    name=cc_name,
    type="amlcompute",
    size="STANDARD_DS3_v2",
    location="westeurope",
    min_instances=0,
    max_instances=2,
    idle_time_before_scale_down=20,
    tier="low_priority",
)

ml_client.begin_create_or_update(cc).result()
````

````python
poller = ml_client.compute.begin_delete("cpu-cluster")
````
