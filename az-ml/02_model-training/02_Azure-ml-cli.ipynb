## Create a compute through the CLI

Create a compute instance through the CLI. 

```python
%%sh

az ml compute create --name cli-created-compute \
                     --size STANDARD_DS3_v2 \
                     --type ComputeInstance \
                     --resource-group $AZURE_RESOURCE_GROUP \
                     --workspace-name $AZURE_ML \
    | grep -E "\"name\"|\"created_on\"|\"state\"|\"type\""
```

Create a compute instance through the CLI based on a YAML file (IaC)

```python
%%sh

echo $(pwd)

az ml compute create --file ./assets/aml-cluster-cli.yml \
                     --resource-group $AZURE_RESOURCE_GROUP \
                     --workspace-name $AZURE_ML \
    | grep -E "\"name\"|\"created_on\"|\"state\"|\"type\""
```

```python
%%sh

az ml compute list --resource-group $AZURE_RESOURCE_GROUP --workspace-name $AZURE_ML | grep \"name\"
```

```python
%%sh 

az ml compute delete --name cli-created-compute --yes \
                     --resource-group $AZURE_RESOURCE_GROUP \
                     --workspace-name $AZURE_ML

az ml compute delete --name aml-cluster --yes \
                     --resource-group $AZURE_RESOURCE_GROUP \
                     --workspace-name $AZURE_ML
```
