# Python-SDK

```python
from azure.ai.ml import MLClient
from azure.identity import DefaultAzureCredential
import dotenv
import os
import json
from pprint import pprint

dotenv.load_dotenv(override=True)
```

```python
ml_client = MLClient(
     credential             = DefaultAzureCredential()
    ,subscription_id        = os.environ["AZURE_SUBSCRIPTION_ID"]
    ,resource_group_name    = os.environ["AZURE_RESOURCE_GROUP"]    
    ,workspace_name         = os.environ["AZURE_ML"]
)
```

## Working with the ML client

### Check some stuff

```python
connections = [conn for conn in ml_client.connections.list()]

json.dumps(connections, indent=4)
```

```python
datastores = [ds for ds in ml_client.datastores.list()]
pprint(datastores)
```

```python
data = [d for d in ml_client.data.list()]
pprint(data)
```

### Compute

```python
from azure.ai.ml.entities import ComputeInstance

ci_name = "SDK-created-compute"
ci_size = "Standard_DS11_v2"


ci_config = ComputeInstance(
     name = ci_name
    ,size = ci_size
    ,idle_time_before_shutdown_minutes=30
)

try: 
    poller = ml_client.compute.begin_create_or_update(
        ci_config
    )

    poller.wait()
    
except Exception as e:
    raise e
```

```python
computes = [c for c in ml_client.compute.list()]
pprint(computes)
```

```python
try:
    ml_client.compute.get(ci_name)
except Exception as e:
    print(f"Compute instance {ci_name} does not exist")
```

```python
poller = ml_client.compute.begin_delete(ci_name)
```

```python
computes = [c for c in ml_client.compute.list()]
pprint(computes)
```
