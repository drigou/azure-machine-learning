# Create an online endpoint

```python
from azure.ai.ml import MLClient
from azure.identity import DefaultAzureCredential
import dotenv
import os

dotenv.load_dotenv(override=True)

ml_client = MLClient(
     credential         = DefaultAzureCredential()
    ,subscription_id    = os.environ["AZURE_SUBSCRIPTION_ID"]
    ,resource_group_name= os.environ["AZURE_RESOURCE_GROUP"]
    ,workspace_name     = os.environ["AZURE_ML"]
)
```

```python
from azure.ai.ml.entities import ManagedOnlineEndpoint

endpoint = ManagedOnlineEndpoint(
     name = "endpoint-diabetes-example"
    ,description = "Online Endpoint For Diabetes Prediction"
    ,auth_mode = "key"
)

ml_client.begin_create_or_update(endpoint).result()
```

```python
from azure.ai.ml.entities import Model, ManagedOnlineDeployment
from azure.ai.ml.constants import AssetTypes 

blue_deployment = ManagedOnlineDeployment(
     name = "blue"
    ,endpoint_name= "endpoint-diabetes-example"
    ,model = "azureml:diabetes-model:2"
    ,environment = "diabetic_prediction_environment_mlflow_skinny:0.0.2"
    ,instance_type = "Standard_F2s_v2"
    ,instance_count = 1
)

ml_client.online_deployments.begin_create_or_update(blue_deployment).result()
```

```python
ml_client.online_endpoints.begin_delete(name="endpoint-diabetes-example")
```
